```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
dataset = convert(ds)
dataset.head(100)
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
```

#4. Summary Statistics of Data

Data analysts and scientists may quickly check dataset completion using this tool. They may rapidly replace missing data to assure model and analytical correctness.

Since the Data is clean we move to the visualisation part .

#6. Show the target variable's distribution graphically.


```python

plt.figure(figsize=(8, 6))

sns.countplot(x='class', data=df, palette='pastel')

plt.title('Class Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```

#7. Brief Idea of Data parts using histograsm


```python
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(10, 15))
axs = axs.flatten()  # Flatten the array of axes to easily iterate over it

# Iterating over the DataFrame columns to plot histograms
for i, col in enumerate(df.columns):
    axs[i].hist(df[col], bins=10, color='skyblue', edgecolor='black')
    axs[i].set_title(col)

# This ensures we do not have an IndexError
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

#8. Correlation matrix
Various linkages are shown by seeing the correlation matrix. Data interpretation, model building, and feature selection are all affected by the direction and intensity of the association. By highlighting the most crucial aspects and their interplay, the heatmap with annotated correlation values facilitates data analysis and decision-making.


```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

#**9. Scatter Plot for relationship**
This boxplot shows the correlation between plasma glucose concentration and the total number of pregnancies. Crucial for medical and health-related investigations, it helps detect any patterns or changes in glucose levels over various pregnancy counts, which aids in understanding the influence of pregnancy on glucose levels.

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='BMI', y='Serum Ins', data=df, color='green', alpha=0.6)
plt.title('BMI vs. Serum Insulin Levels')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('2-Hour Serum Insulin (mu U/ml)')
plt.show()
```

#**10. Boxplot for pg concentration and Pragnancies**
This boxplot shows the correlation between plasma glucose concentration and the total number of pregnancies. It allows for the detection of any differences or trends in glucose levels throughout different gestational ages. Gaining insight into these correlations is critical for health-related and medical analysis, as it helps to understand the impact of pregnancy on glucose concentrations.







```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pregnancies', y='PG Concentration', data=df, palette='coolwarm')
plt.title('PG Concentration by Number of Pregnancies')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Plasma Glucose Concentration')
plt.show()
```
